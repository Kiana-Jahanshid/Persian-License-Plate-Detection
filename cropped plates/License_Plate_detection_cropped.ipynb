# CROP License-plate

To crop the license plate region from the image and pass it to an OCR system, you can follow these steps:

After detecting the license plate region using your model, obtain the coordinates of the bounding box that surrounds the plate.

Use the coordinates to crop the license plate region from the original image. You can use a library like OpenCV to perform this operation. Here's some example code:

```python
bounding_boxes = []
counter=0
for image in detected_images_list:
    detections = new_model.track(image, persist=True)[0]
    detections.save_crop('outputs')
    # print(results)

    for detection in detections.boxes.data.tolist():
        print(detection)
        if len(detection)==7:
          x1, y1, x2, y2, track_id, score, class_id = detection
        elif len(detection)==6:
          x1, y1, x2, y2, track_id, score = detection

        if  score > 0.1:
            bounding_boxes.append([x1, y1, x2, y2][counter])
    counter+=1
for i in range(len(bounding_boxes)) :
  print(bounding_boxes[i])
```
